```python
# Import Dependencies
from splinter import Browser
from bs4 import BeautifulSoup
from pprint import pprint

import pandas as pd
```

```python
# Create a Chrome browser instance
browser = Browser('chrome')
```

```python
# Visit the page and create the soup object
url = 'https://www.sports-reference.com/cfb/schools/'
browser.visit(url)

# Extract objects from the webpage
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
tables = soup.find_all('tbody')

# Inspect the html
pprint(tables)
```

```python
# Extract out the links from the table (tables[4] table in the HTML)
# Create an empty list
school_list = []
link_list = []

# Iterate through the rows in the table (table[4] contains the correct table)
for row in tables[1].find_all('tr'):
    
    # Find the first <td> cell with the class "left"
    td_cell = row.find('td', {'class': 'left'})
    
    if td_cell:
        # Find the <a> tag within the <td> cell
        link_cell = td_cell.find('a')
        
        if link_cell:
            # Extract the href and link text
            href = link_cell['href']
            link_text = link_cell.get_text()

            # Append the href and link text to the data_list
            school_list.append(link_text)
            link_list.append(href)

# Create DataFrame and create the .csv file
df = pd.DataFrame({'school': school_list, 'link': link_list})
df.to_csv('resources/schools.csv', index=False)
```
